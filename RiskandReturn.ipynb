```python
import numpy as np 
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
MNKD = wb.DataReader('MNKD', data_source="yahoo", start='2012-1-1')
```

```python
MNKD.tail(20)
```

```python
# Calculate the simple rate of return 
MNKD['simple_return'] = (MNKD['Adj Close'] / MNKD['Adj Close'].shift(1)) - 1
print(MNKD['simple_return'])
```

```python
# plot the daily returns 
MNKD['simple_return'].plot(figsize=(8,5)) 
plt.show()
```

```python
# Calculate the average daily return
avg_returns_d = MNKD['simple_return'].mean()
avg_returns_d
```

```python
# Calculate the average annual return per 250 trading days
avg_returns_a = MNKD['simple_return'].mean() * 250
print(str(round(avg_returns_a,5)*100) + '%')
```

```python
# Calclulate the log returns for a given security
MNKD['log_return'] = np.log(MNKD['Adj Close'] / MNKD['Adj Close'].shift(1))
print(MNKD['log_return'])
```

```python
# Plot the log returns
MNKD['log_return'].plot(figsize=(8,5))
plt.show()
```

```python
# find the log daily return
log_return_d = MNKD['log_return'].mean()
# find the log annual return
log_return_a = MNKD['log_return'].mean() * 250
print(str(round(log_return_a,5) * 100) + '%')
```
