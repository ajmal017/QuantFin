```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers = ['MNKD', 'ACRX', 'ECYT']
portfolio = pd.DataFrame()

for t in tickers:
    portfolio[t] = wb.DataReader(t, data_source='yahoo', start='2017-1-1')['Adj Close']
    
```

```python
portfolio.tail()
```

```python
portfolio_returns = np.log(portfolio / portfolio.shift(1))
```

```python
portfolio_returns
```

```python
# Calculate the mean of security daily returns
MNKD_Mean_d = portfolio_returns['MNKD'].mean()
ECYT_Mean_d = portfolio_returns['ECYT'].mean()
ACRX_Mean_d = portfolio_returns['ACRX'].mean()
```

```python
# Calculate the mean of security annual returns
MNKD_Mean_a = portfolio_returns['MNKD'].mean() * 250
ECYT_Mean_a = portfolio_returns['ECYT'].mean() * 250
ACRX_Mean_a = portfolio_returns['ACRX'].mean() * 250
```

```python
# Calculate the standard deviation of daily security returns
MNKD_STD_d = portfolio_returns['MNKD'].std()
ECYT_STD_d = portfolio_returns['ECYT'].std()
ACRX_STD_d = portfolio_returns['ACRX'].std()
```

```python
# Calculate the standard deviation of annual security returns
MNKD_STD_d = portfolio_returns['MNKD'].std() * 250 ** 0.5
ECYT_STD_d = portfolio_returns['ECYT'].std() * 250 ** 0.5
ACRX_STD_d = portfolio_returns['ACRX'].std() * 250 ** 0.5
```
